```python
df = pd.read_csv('EDL_Train.csv')
df.columns = df.columns.str.replace(r'train_data_t2p.', '')
df['customer_nbr'] = df['customer_nbr'].astype(str)
df['invoice_nbr'] = df['invoice_nbr'].astype(str)
df["ratio_closed_0"] = df["closed_bucket_0"]/df["count_paid_invoices"]
df["ratio_closed_1"] = df["closed_bucket_1"]/df["count_paid_invoices"]
df["ratio_closed_2"] = df["closed_bucket_2"]/df["count_paid_invoices"]
df["ratio_closed_3"] = df["closed_bucket_3"]/df["count_paid_invoices"]
df["ratio_closed_4"] = df["closed_bucket_4"]/df["count_paid_invoices"]
```

```python
df.head()
```

```python
def cc(row):
    if row['country_code']==1:
        return 'US'
    else:
        return 'CA'
    
df['country_code'] =df.apply(lambda row: cc(row), axis=1)
```

```python
df.country_code.value_counts()
```

```python
def it(row):
    if row['invoice_oks']==1:
        return 'invoice_oks'
    elif row['tm_invoice']==1:
        return 'tm_invoice'
    elif row['om_invoice']==1:
        return 'om_invoice'
    elif row['hsr_sca_invoice']==1:
        return 'hsr_sca_invoice'
    elif row['sca_invoice']==1:
        return 'sca_invoice'
    elif row['saas_oks_invoice']==1:
        return 'saas_oks_invoice'
    elif row['hsr_tm_invoice']==1:
        return 'hsr_tm_invoice'
    elif row['hsr_invoice_oks']==1:
        return 'hsr_invoice_oks'
    elif row['hsr_om_invoice']==1:
        return 'hsr_om_invoice'
    elif row['other_invoice_type_name']==1:
        return 'other_invoice_type'
```

```python
df['invoice_type'] =df.apply(lambda row: it(row), axis=1)
```

```python
df.invoice_type.value_counts()
```

```python
def ind(row):
    if row['financial_banking_cfi']==1:
        return 'financial_banking_cfi'
    elif row['financial_banking_major']==1:
        return 'financial_banking_major'
    elif row['financial_channel']==1:
        return 'financial_channel'
    elif row['financial_other_financial_serv']==1:
        return 'financial_other_financial_serv'
    elif row['hospitality_channel']==1:
        return 'hospitality_channel'
    elif row['hospitality_eating_drinking_qs']==1:
        return 'hospitality_eating_drinking_qs'
    elif row['hospitality_eating_drinking_ts']==1:
        return 'hospitality_eating_drinking_ts'
    elif row['hospitality_sports_recreation']==1:
        return 'hospitality_sports_recreation'
    elif row['retail_channel']==1:
        return 'retail_channel'
    elif row['retail_cstore_and_petroleum']==1:
        return 'retail_cstore_and_petroleum'
    elif row['retail_distribution_wholesale']==1:
        return 'retail_distribution_wholesale'
    elif row['retail_drug']==1:
        return 'retail_drug'
    elif row['retail_food_and_mass']==1:
        return 'retail_food_and_mass'
    elif row['retail_public_sector']==1:
        return 'retail_public_sector'
    elif row['retail_specialty_retail']==1:
        return 'retail_specialty_retail'
    elif row['t_and_t_technology']==1:
        return 't_and_t_technology'
    elif row['t_and_t_telecom']==1:
        return 't_and_t_telecom'
    elif row['tlg_channel']==1:
        return 'tlg_channel'
    elif row['tlg_gaming']==1:
        return 'tlg_gaming'
    elif row['tlg_lodging']==1:
        return 'tlg_lodging'
    elif row['tlg_travel']==1:
        return 'tlg_travel'
    elif row['other_industry']==1:
        return 'other_industry'
    
df['industry'] =df.apply(lambda row: ind(row), axis=1)
```

```python
df.industry.value_counts()
```

```python
df.head()
```

```python
def cs(row):
    if row['customer_segment']==1:
        return 'Less Value Customer'
    else:
        return 'High Value Customer'
    
df['customer_segment'] =df.apply(lambda row: cs(row), axis=1)
```
